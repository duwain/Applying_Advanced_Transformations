```python
df = pd.read_csv('superhero_info - superhero_info.csv')
df.head()
```

```python
df.info()
```

```python
df['Measurements'].str.split(' ',expand=True)
```

```python
df[['h','Height_cm','cm', 'w','weight_kg', 'kg']] = df['Measurements'].str.split(' ',expand=True)
df.head(2)
```

```python
df = df.drop('Measurements', axis=1)
df = df.drop('h', axis=1)
df = df.drop('w', axis=1)
df = df.drop('cm', axis=1)
df = df.drop('kg', axis=1)
```

```python
df.head(2)
```

```python
df['Height_cm'] = df.Height_cm.str.extract(r'(\d+)', expand=True).astype(int)
```

```python
df.head(2)
```

```python
df['weight_kg'] = df.weight_kg.str.extract(r'(\d+)', expand=True).astype(int)
```

```python
df.head(2)
```

```python
df[['Hero','Publisher']] = df['Hero|Publisher'].str.split('|',expand=True)
df.head(2)
```

```python
df = df.drop('Hero|Publisher', axis=1)
df.head(2)
```

```python
df.info()
```

```python
df_powers['Power_split'] = df_powers.Powers.str.split(',')
```

```python

df_powers.head(2)
```

```python
df_powers['Power_split'].value_counts()
```

```python
exploded = df_powers.explode('Power_split')
exploded[['hero_names','Powers','Power_split']].head(5)
```

```python
## saving the unique values from the exploded column
cols_to_make = exploded['Power_split'].dropna().unique()
cols_to_make

```

```python
for col in cols_to_make:
    df_powers[col] = df_powers['Powers'].str.contains(col)
df_powers.head()
```

```python
df_powers = df_powers.drop(columns=['Powers','Power_split'])
```

```python
df_powers.head(2)
```

```python
df_powers.rename(columns = {'hero_names':'Hero'}, inplace = True)
```
